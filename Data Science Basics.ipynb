# Poisson Distribution Using Python

```python
# Problem Statement

# Suppose you are going for a long drive.The rate of occurences of good restaurants in a range of 10 miles is 2.
# In other words, the mean number of occurences of restuarants in a range of 10 miles is 2.
# What is the probability that 0, 1, 2, 3, 4 or 5 restuarants will show up in the next 10 miles.
```

```python
# Scipy.stats poisson class is used along with pmf method to calculate the value of probabilities
from scipy.stats import poisson
import matplotlib.pyplot as plt

# Random variable representing number of restuarants.
# Mean number of occurences of restuarants in 10 miles is 2.

X = [0, 1, 2, 3, 4, 5]
lmbda = 2
```

```python
# Probability values
poisson_pd = poisson.pmf(X, lmbda)
```

```python
# Plot the probability distribution
fig, ax = plt.subplots(1, 1, figsize=(8,6))
ax.plot(X, poisson_pd, 'bo', ms=8, label='poisson pmf')
plt.ylabel("Probability", fontsize="18")
plt.xlabel("X = No. of Restuarants", fontsize="18")
plt.title("Poisson Distribution - No. of Restaurants Vs Probability", fontsize="18")
ax.vlines(X, 0, poisson_pd, colors='b', lw=5, alpha=0.5)
```

```python
import random 
random.seed(20)

x = np.array([random.gauss(100,15) for x in range(20)])
y = np.array([random.gauss(100,15) for x in range(20)])

f_test_stat = np.var(x,ddof=1) / np.var(y,ddof = 1)

dfn = x.size - 1
dfd = y.size - 1

p_value = 1 - f.cdf(f_test_stat,dfn,dfd)
p_value
```
